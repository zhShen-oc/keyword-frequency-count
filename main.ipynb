```python
import os
import jieba
```

第一步：文本文件的读取

```python
data_path = './textdata/'
file_list = []
res_path = './B1_result/'
res_list =  []
for first_dir in os.listdir(data_path):
    if not os.path.exists(res_path + first_dir):
        os.mkdir(res_path + first_dir)
    for second_dir in os.listdir(data_path + first_dir):
        if not os.path.exists(res_path + first_dir + '/' + second_dir):
            os.mkdir(res_path + first_dir + '/' + second_dir)
        for file in os.listdir(data_path + first_dir + '/' + second_dir):
            file_list.append(data_path + first_dir + '/' + second_dir + '/' + file)
            res_list.append(res_path + first_dir + '/' + second_dir + '/' + file)
        
```

第二步：中文词库分词

```python
def split_text(text, mode):
    txt = open(text, "r", encoding="utf-8",errors='ignore').read().strip()
    if mode == 'cut_all': #全模式
        seg_list = jieba.cut(txt, cut_all=True)
    elif mode == 'cut_word': #精确模式
        seg_list = jieba.cut(txt, cut_all=False)
    else:
        seg_list = jieba.cut(txt, cut_all=False)
    # print(" ".join(seg_list))
    return seg_list
```

第三步：获取关键词

```python
keywords_data_path = './keywords/'

def get_keywords(keywords_name):
    keywords_txt = open(os.path.join(keywords_data_path, keywords_name+'.txt')
    , "r", encoding="ANSI").read().strip()
    keywords = keywords_txt.split('|')
    print("指定的关键词如下：")
    print(keywords)
    return keywords
```

第四步：统计分词后的文本关键词频数

```python
def count_frequency(seg_list, keywords):
    frequency_dict = {}
    for keyword in keywords:
        frequency_dict[keyword] = 0
    for word in seg_list:
        if word in frequency_dict:
            frequency_dict[word] += 1
    return frequency_dict
    
```

第五步：输出结果

```python
import xlsxwriter
def main(keywords_name, mode):

    print("获取关键词！")
    keywords = get_keywords(keywords_name)
    #将关键词加入jieba词库
    for keyword in keywords:
        jieba.add_word(keyword)
    print('关键词加入成功！')
    for i, file in enumerate(file_list):
        seg_list = split_text(file, mode=mode)
        frequency_dict = count_frequency(seg_list, keywords)
        # open the file in the write mode
        filename = res_list[i][:-4]+'.xlsx'
        # Create an new Excel file and add a worksheet.
        with xlsxwriter.Workbook(filename) as workbook:
          # Add worksheet
            worksheet = workbook.add_worksheet()
            i = 0
            for k, v in frequency_dict.items():
                worksheet.write(i, 0, k)
                worksheet.write(i, 1, v)
                i += 1
        print("文件" + filename + "处理完成！")
    print("全部处理完成！")
```

```python
# 从 ['A1','A2','A3','A4','A5','B1','B2','B3','B4','B5'] 选一个系列的关键词
main('B1', mode = 'cut_all')
```
